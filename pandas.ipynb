# pd.agg

# Parameters

func: Function(s) to apply. It can be:

- A single function (e.g., 'sum', 'mean').
- A list of functions (e.g., ['sum', 'mean']).
- A dictionary mapping column names to functions (e.g., {'col1': 'sum', 'col2': ['mean', 'max']}).
- A custom Python function.
- axis: Whether to apply the function along rows (1) or columns (0, default).
- *args and **kwargs: Additional arguments to pass to the

```python
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

result = df.agg('sum')
print(result)
```

```python
result = df.agg(['sum', 'mean'])
print(result)
```

```python
result = df.agg(lambda x: x.max() - x.min())
print(result)
```

```python
result = df.agg({'A': 'sum', 'B': ['mean', 'max']})
print(result)
```

```python
s = pd.Series([1, 2, 3])

result = s.agg(['sum', 'mean'])
print(result)
```

axis Parameter:

axis=0 (default): Apply functions along columns.
axis=1: Apply functions along rows.

```python
result = df.agg('sum', axis=1)
print(result)
```

## Date Format Codes:

- %Y - Year with century (e.g., 2025)
- %y - Year without century (00-99) (e.g., 25)
- %m - Month as a zero-padded number (01-12) (e.g., 01)
- %B - Full month name (e.g., January)
- %b - Abbreviated month name (e.g., Jan)
- %d - Day of the month as a zero-padded number (01-31) (e.g., 09)
- %j - Day of the year (001-366) (e.g., 343)
- %w - Weekday as a number (0=Sunday, 6=Saturday) (e.g., 1 for Monday)
- %A - Full weekday name (e.g., Monday)
- %a - Abbreviated weekday name (e.g., Mon)
- %U - Week number of the year (Sunday as the first day of the week) (e.g., 49)
- %W - Week number of the year (Monday as the first day of the week) (e.g., 50)
## Time Format Codes:

- %H - Hour (00-23) (e.g., 15)
- %I - Hour (01-12, 12-hour clock) (e.g., 03)
- %p - AM or PM (e.g., PM)
- %M - Minute (00-59) (e.g., 45)
- %S - Second (00-59) (e.g., 08)
- %f - Microsecond (000000-999999) (e.g., 123456)
## Combined Date and Time Codes:

- %c - Full date and time (e.g., Mon Dec 9 15:45:08 2013)
- %x - Locale’s appropriate date (e.g., 12/09/13 in US, 09.12.2013 in EU)
- %X - Locale’s appropriate time (e.g., 15:45:08)
## Timezone Codes:

- %z - UTC offset in ±HHMM format (e.g., +0000, -0500)
- %Z - Timezone name (e.g., UTC, EST)
- Miscellaneous Codes:

%% - A literal % character (e.g., %)

```python
import pandas as pd
from datetime import datetime

# Create the DataFrame
df = pd.DataFrame({
    'name': ['ram', 'allu arjun', 'nani'],
    'DOB': ['01-10-2000', '01-12-1985', '02-12-1984']
})

# Convert the 'DOB' column to datetime format
df['DOB'] = pd.to_datetime(df['DOB'], format='%d-%m-%Y')

# Calculate age by subtracting DOB from today's date
today = datetime.today()
df['age'] = df['DOB'].apply(lambda dob: today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day)))

print(df)
```

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'name': ['ram', 'allu arjun', 'nani'],
    'DOB': ['01-10-2000', '01-12-1985', '02-12-1984']
})

# Convert DOB column to datetime
df['DOB'] = pd.to_datetime(df['DOB'], format='%d-%m-%Y')

# Let's assume we are comparing the DOBs to a specific date
specific_date = pd.to_datetime('01-01-2025', format='%d-%m-%Y')

# Calculate the difference between the specific date and DOB
df['days_difference'] = specific_date - df['DOB']

# Show the result
print(df)
```
