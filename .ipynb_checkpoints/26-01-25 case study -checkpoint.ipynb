2. Write a lambda function that returns 'In Stock' if the stock is greater than 100,    otherwise returns 'Less Stock' in the new column Stock_category


```python
#a. using for loop
res=['In Stock' if i>100 else 'Less Stock' for i in df['Stock']]
res
```

```python
#b. using direct approach (boolean indexing)
np.where(df['Stock']>100,"In Stock",'Less Stock') 
```

```python
#c. using apply and user defined function with def
def v(x):
    if x>100:
        return 'In Stock'
    else:
        return 'Less Stock'
df['Stock'].apply(v) 
```

```python
#d. using apply and lambda 
df['Stock'].apply(lambda x:'In Stock' if x>100 else 'Less Stock') 
```

```python
df['Stock'].map(lambda x:'In Stock' if x>100 else 'Less Stock')
```

```python
df['Stock_category']=df['Stock'].map(lambda x:'In Stock' if x>100 else 'Less Stock')
df
```

3. create a temporary column when price is >100 and product is electronics give value as 1 when product is accessories give values as 1 otherwise it should be 0.


```python
#a. using for loop
res=[]
for i,j in list(zip(df['Price'],df['Category'])):
    if (i>100) and (j==('Electronics')or('Accessories')):
        res.append(1)
       
    else:
        res.append(0)
res
```

```python
#b. using direct approach (boolean indexing)
df["temp"] = 0
df.loc[(df["Price"] > 100) & (df["Category"] == "Electronics"),"temp"] = 1
df.loc[(df["Price"] > 100) & (df["Category"] == "Accessories"),"temp"] = 1
df
```

```python
#c. using apply and user defined function with d
def assign_temp(row):
    if row["Price"] > 100 and row["Category"] == "Electronics":
        return 1
    elif row["Price"] > 100 and row["Category"] == "Accessories":
        return 1
    else:
        return 0

df.apply(assign_temp, axis=1)
```

```python
#d. using apply and lambda 
df.apply(lambda x: 1 if x['Price']>100 and x['Category'] in ['Electronics','Accessories'] else 0,axis=1)
```

```python
#e. using map
list(map(lambda x: 1 if x[0]>100 and x[1] in ['Electronics','Accessories'] else 0,zip(df['Price'],df['Category'])))
```
