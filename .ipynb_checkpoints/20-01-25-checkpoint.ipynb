```python
import pandas as pd
df=pd.read_csv(r"C:\Users\dtrri\bot\Salary Data.csv")
df
```

```python
# What is the mean, median, and standard deviation of the YearsExperience and Salary columns?
df.describe()
```

```python
# Find the minimum and maximum salary in the dataset.
print(f'minimum salary : {df['Salary'].min()}\nmaximum salary : {df['Salary'].max()}')
```

```python
#Retrieve all employees with more than 5 years of experience.
df[df['YearsExperience']>5]
```

```python
#Identify employees whose salary is below the average salary.
df[df['Salary']<df['Salary'].mean()]
```

```python
#Find the top 5 employees with the highest salaries.
df.sort_values(by='Salary', ascending=False).head(5)
```

```python
#Sort the dataset in descending order of experience.
df.sort_values(by='YearsExperience', ascending=False)
```

```python
#Find the total number of employees who have less than 3 years of experience.
df[df['YearsExperience']<3]
print(len(df[df['YearsExperience']<3]))
```

```python
#How many employees have exactly 5 years of experience?
print(len(df[df['YearsExperience']==5]))
```
