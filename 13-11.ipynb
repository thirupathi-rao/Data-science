```python
Q4. Check if the string s (s = "madam") is a palindrome and assign the result (True/False) to the
variable is_palindrome.
```

```python
def is_palindrome(str):
        for i in range(0, int(len(str)/2)):
            if str[i] != str[len(str)-i-1]:
                return False
        return True
s = "madam"
x=is_palindrome(s)
if x:
    print('true')
else:
    print('false')
```
